```python
import pandas as pd

filepath = '/scratch/nvu5gw/Project 3/data/nacti_metadata.csv'

df = pd.read_csv(filepath, low_memory = False)

print(f"Total data rows: {len(df)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Distribution of common_name
plt.figure(figsize=(10, 6))
top_names = df['common_name'].value_counts().head(20)
top_names.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Top 20 Common Names')
plt.xlabel('Common Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("Animal Distribution")
plt.show()
```

```python
# 2. Distribution of location
plt.figure(figsize=(10, 6))
top_locations = df['location'].value_counts().head(20)
top_locations.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Distribution of Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("Location Distribution")
plt.show()
```

```python
# 3. Distribution of common_name by location
top_locs = df['location'].value_counts().head(10).index
top_names = df['common_name'].value_counts().head(20).index
filtered_df = df[df['location'].isin(top_locs) & df['common_name'].isin(top_names)]

crosstab_df = pd.crosstab(filtered_df['location'], filtered_df['common_name'])
crosstab_df.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')
plt.title('Top 20 Animals by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.legend(title='Common Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("Animals by Location")
plt.show()
```
